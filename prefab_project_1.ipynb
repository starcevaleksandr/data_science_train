# Исследование закономерностей, определяющих успешность игры.

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

У нас есть данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

#### Преобразование типов данных в столбцах и работа с пропусками в данных.

```python
data.isna().sum()
```

Избавимся от пропусков в столбце name и genre, удалив их из датасета, т.к. 2 записи в датасете не повлияют на результат анализа.

```python
data.query('name.isna()')
```

```python
data = data.query('not name.isna()')
```

```python
data.query('year_of_release.isna()')
```

Закономерностей в пропуске года выпуска игры не видно, а количество пропусков мало, в сравнении с размером датасета. Поэтому от пропущенных данных можно избавиться, затем преобразовать `year_of_release` в int, т.к. годы должны иметь целочисленный тип данных, но имеют дробную часть .0, от этого необходимо избавиться.

```python
data = data.query('not year_of_release.isna()')
data['year_of_release'] = data['year_of_release'].astype('int16')
```

```python
#проверка целостности данных
data.info()
data['year_of_release'].isna().sum()
```

Теперь преобразуем типы данных `object` в численные, где это возможно. Обратим внимание на столбец `user_score`. В нем присутствуют значения 'tbd', всё остальное - числа и NaN.

```python
data.query('user_score.isna()')
```

```python
data.query('user_score == "tbd"')
```

Возможный вариант значения tbd:
* TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Акроним служит для отметки неясностей или пропусков, которые надо заполнить, в информации требований.

Так это или нет, сложно однозначно сказать. В любом случае, заполнить такие данные числами мы не сможем, поэтому будем относить эти данные к пропускам. 

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
```

```python
data['user_score'] = data['user_score'].astype('float')
```

```python
data.isna().sum()
```

```python
data.query('user_score.isna()')
```

Как видно, многие пропуски находятся в столбцах `critic_score`, `user_score`, `rating` одновременно, а также эти пропуски встречаются в огромных количествах. Возможная причина появления: отсутствие оценки этих игр на популярных ресурсах. Такие пропуски нельзя заполнить маркерными данными или же медианными значениями, иначе заполнение может повлиять на характер связи между заполняемой величиной и остальными.

В столбце `rating` значения NaN следует заменить маркерными значениями, например Unknown, это пригодится для дальнейшего анализа данных.


```python
data['rating'] = data['rating'].fillna('unknown')
```

#### Как менялись продажи по платформам.

Посчитаем количество продаж по каждой платформе. Построим графики продаж в каждый год по первым 10 платформам. 

```python
platform_sales = \
    data.pivot_table(index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales', ascending=False)

platform_sales.head(10)
```

```python
for i in platform_sales.head(10).index:
    data.query('platform == @i').groupby('year_of_release').agg({'total_sales': 'sum'}).plot(y='total_sales', grid=True, legend=True, label=i)
```

В основном платформы за 5-8 лет проходят период пик активного выпуска игр. Так, например, у PS2 - с 2000 по 2006 год наибольшее количество продаж. У X360 с 2007 по 2013.

#### Актуальный период для прогноза.

Т.к. ранее было замечено, что жизненный цикл платформ около 5-8 лет, а данные до 2001 решено было вообще не учитывать, определим актуальный период для построения прогноза на 2017 год.

На мой взгляд, будет актуально использовать данные с 2013 года по 2016. Т.к. пики продаж в 2008 и 2009 уже пройдены, а сроки актуальности платформ как раз с запасом впишутся в этот период. К тому же, вряд ли будет очень много продаж слишком давно выпущенных игр.

### Проверка гипотез.

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

* H0: Средние пользовательские рейтинги платформ Xbox One и PC равны;
* H1: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Гипотеза о равенстве средних двух генеральных совокупностей

```python
data_actual.head()
```

```python
xone_data = data_actual.query('platform == "XOne" and not user_score.isna()')['user_score']
pc_data = data_actual.query('platform == "PC" and not user_score.isna()')['user_score']
print('Средний пользовательский рейтинг для Xbox One:', xone_data.mean())
print('Средний пользовательский рейтинг для ПК:', pc_data.mean())
```

Проверим гипотезу с помощью t-теста. Уровень значимости - 5%.

```python
alpha = 0.05
result = st.ttest_ind(xone_data, pc_data, equal_var=True)
print('p-value:', result.pvalue)
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: 
    print('Не получилось отвергнуть нулевую гипотезу')
```

Не получилось отвергунть нулевую гипотезу о равенстве средних пользовательских рейтингах для Xbox One и ПК. Поэтому считаем средние пользовательские рейтинги для этих платформ одинаковыми.

#### Средние пользовательские рейтинги жанров Action и Sports разные.

* H0: Средние пользовательские рейтинги жанров Action и Sports равны.
* H1: Средние пользовательские рейтинги жанров Action и Sports не равны.

Гипотеза о равенстве средних двух генеральных совокупностей.

```python
action_data = data_actual.query('genre == "Action" and not user_score.isna()')['user_score']
sports_data = data_actual.query('genre == "Sports" and not user_score.isna()')['user_score']
print('Средний пользовательский рейтинг для жанра Action:', action_data.mean())
print('Средний пользовательский рейтинг для жанра Sports:', sports_data.mean())
```

Проверим гипотезу с помощью t-теста. Уровень значимости - 5%.

```python
alpha = 0.05
result = st.ttest_ind(action_data, sports_data, equal_var=True)
print('p-value:', result.pvalue)
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: 
    print('Не получилось отвергнуть нулевую гипотезу')
```

Отвергаем гипотезу о равенстве средних оценок пользователей по жанрам Action и Sports. А значит, средние оценки пользователей по этим жанрам действительно разные.
